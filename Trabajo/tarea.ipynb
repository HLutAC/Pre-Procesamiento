```python
import pandas as pd
data = pd.read_csv("consumo_202307.csv", encoding="latin-1", delimiter=";")
data.head()
```

```python
print(data.isnull().sum())
```

```python
print(data.dtypes)
data.head()
```

```python
data["UUID"]=data["UUID"].astype('category')
data["DEPARTAMENTO"]=data["DEPARTAMENTO"].astype('category')
data["PROVINCIA"]=data["PROVINCIA"].astype('category')
data["DISTRITO"]=data["DISTRITO"].astype('category')
data["LOCALIDAD"]=data["LOCALIDAD"].astype('category')
data["TARIFA"]=data["TARIFA"].astype('category')
data["FASE"]=data["FASE"].astype('category')
data["CODIGO_CIU"]=data["CODIGO_CIU"].astype('category')
print(data.dtypes)
data.head()
```

```python
categorical_data=data.drop(['FECHA_CORTE', 'PERIODO_FACTURACION', 'UBIGEO', 'FECHA_INICIO','CONSUMO_KW','POTENCIA_CONTRATADA', 'MONTO_SOLES'],axis=1)
categorical_data.head()
```

```python
categorical_data_label=categorical_data
```

```python
categorical_data_label['DEPARTAMENTO']=categorical_data_label['DEPARTAMENTO'].cat.codes
categorical_data_label['PROVINCIA']=categorical_data_label['PROVINCIA'].cat.codes
categorical_data_label['DISTRITO']=categorical_data_label['DISTRITO'].cat.codes
categorical_data_label['LOCALIDAD']=categorical_data_label['LOCALIDAD'].cat.codes
categorical_data_label['TARIFA']=categorical_data_label['TARIFA'].cat.codes
categorical_data_label['FASE']=categorical_data_label['FASE'].cat.codes
categorical_data_label['CODIGO_CIU']=categorical_data_label['CODIGO_CIU'].cat.codes
categorical_data_label
```

```python
numerical_data = data.drop(['UUID','DEPARTAMENTO','PROVINCIA','DISTRITO','LOCALIDAD','TARIFA','FASE','CODIGO_CIU'],axis=1)
numerical_data.head()
```

```python
dataset_new=categorical_data_label.join(numerical_data)
dataset_new.head()
```

```python
data = pd.read_csv("consumo_202307.csv", encoding="Latin-1", delimiter=";")
categorical_data_one_hot = data.drop(['UUID','FECHA_CORTE', 'PERIODO_FACTURACION', 'UBIGEO', 'FECHA_INICIO','CONSUMO_KW','POTENCIA_CONTRATADA', 'MONTO_SOLES'],axis=1)
categorical_data_one_hot["DEPARTAMENTO"]=categorical_data_one_hot["DEPARTAMENTO"].astype('category')
categorical_data_one_hot["PROVINCIA"]   =categorical_data_one_hot["PROVINCIA"].astype('category')
categorical_data_one_hot["DISTRITO"]    =categorical_data_one_hot["DISTRITO"].astype('category')
categorical_data_one_hot["LOCALIDAD"]   =categorical_data_one_hot["LOCALIDAD"].astype('category')
categorical_data_one_hot["TARIFA"]      =categorical_data_one_hot["TARIFA"].astype('category')
categorical_data_one_hot["FASE"]        =categorical_data_one_hot["FASE"].astype('category')
categorical_data_one_hot["CODIGO_CIU"]  =categorical_data_one_hot["CODIGO_CIU"].astype('category')

print(categorical_data_one_hot.dtypes)

encoded_departament = pd.get_dummies(categorical_data_one_hot['DEPARTAMENTO'])
encoded_provincia = pd.get_dummies(categorical_data_one_hot['PROVINCIA'])
encoded_distrito = pd.get_dummies(categorical_data_one_hot['DISTRITO'])
encoded_localidad = pd.get_dummies(categorical_data_one_hot['LOCALIDAD'])
encoded_tarifa = pd.get_dummies(categorical_data_one_hot['TARIFA'])
encoded_fase = pd.get_dummies(categorical_data_one_hot['FASE'])
encoded_codigo_ciu = pd.get_dummies(categorical_data_one_hot['CODIGO_CIU'])


categorical_data_one_hot=categorical_data_one_hot.join(encoded_departament)
categorical_data_one_hot=categorical_data_one_hot.join(encoded_provincia, rsuffix='_provincia')
categorical_data_one_hot=categorical_data_one_hot.join(encoded_distrito, rsuffix='_distrito')
categorical_data_one_hot=categorical_data_one_hot.join(encoded_localidad, rsuffix='_localidad')
categorical_data_one_hot=categorical_data_one_hot.join(encoded_tarifa)
categorical_data_one_hot=categorical_data_one_hot.join(encoded_fase)
categorical_data_one_hot=categorical_data_one_hot.join(encoded_codigo_ciu)
categorical_data_one_hot.head()
#genera un 1GB de CSV (:'v)
#categorical_data_one_hot.to_csv("categorical_data_encoded.csv", index=False)


```
