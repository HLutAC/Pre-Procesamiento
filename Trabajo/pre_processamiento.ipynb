```python
data = pd.read_csv("combinados.csv", encoding="latin-1", delimiter=";")
data = data.dropna()
print(data.isnull().sum())
```

```python
print(data.dtypes)
```

# Categorical Encoding

```python
data["TARIFA"]        = data["TARIFA"].astype('category')
data["CODIGO_CIU"]  = data["CODIGO_CIU"].astype('category')
data["POTENCIA_CONTRATADA"]  = data["POTENCIA_CONTRATADA"].astype('category')
print(data.dtypes)
```

```python
categorical_data=data.drop(['UUID',
                            'CONSUMO_KW_202305','MONTO_SOLES_202305',
                            'CONSUMO_KW_202306','MONTO_SOLES_202306',
                            'CONSUMO_KW_202307','MONTO_SOLES_202307',
                            'CONSUMO_KW_202308','MONTO_SOLES_202308',
                            'CONSUMO_KW_202309','MONTO_SOLES_202309',
                            'UBIGEO', 'DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 'LOCALIDAD'], axis=1)
categorical_data.head(7)
print(categorical_data.dtypes)
```

# Label Encoding

```python
categorical_data["TARIFA"]                  = categorical_data["TARIFA"].cat.codes
categorical_data["CODIGO_CIU"]              = categorical_data["CODIGO_CIU"].cat.codes
categorical_data["POTENCIA_CONTRATADA"]     = categorical_data["POTENCIA_CONTRATADA"].cat.codes
categorical_data.head(1000)
```

```python
numerical_data=data.drop(['UUID', 'TARIFA','FASE','CODIGO_CIU','POTENCIA_CONTRATADA','UBIGEO', 'DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 'LOCALIDAD'], axis=1)

numerical_data.head()
```

```python
dataset_new=categorical_data.join(numerical_data)
dataset_new.head()
```

# One Hot Encoding

```python
encoded_fase = pd.get_dummies(dataset_new['FASE'])

dataset_new = dataset_new.join(encoded_fase)
dataset_new.head()
```

# Target Encoding

```python
dataset_new['CODIGO_CIU'] = dataset_new['CODIGO_CIU'].astype('category')
dataset_new['CODIGO_CIU'] = dataset_new['CODIGO_CIU'].cat.codes
dataset_new.head()
```

```python
codigos_ciu_comunes = dataset_new['CODIGO_CIU'].value_counts()
print(codigos_ciu_comunes)
```

```python
import matplotlib.pyplot as plt
codigos_ciu_comunes = dataset_new['CODIGO_CIU'].value_counts()


plt.figure(figsize=(10, 6))

codigos_ciu_comunes.plot(kind='bar')

plt.xlabel('Código CIU')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Códigos CIU')

plt.show()
print(codigos_ciu_comunes)
categorical_data.head()
```
